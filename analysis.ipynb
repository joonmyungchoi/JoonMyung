```python
import torch
import os

# 0.Setting
# Section A. View Point
from joonmyung.draw import drawImgPlot
from playground.analysis.dataset import JDataset
from playground.analysis.model import JModel

view = [True, False, False, True, True] # [Image, Attention, Rollout, GradCam, What I Want]
save_path, save = '/hub_data/joonmyung/conference/playgrounds/results/', False

# Section B. Data
root_path, dataset = "/hub_data/joonmyung/data", "imagenet"
data_num = [[0, 0], [1, 0], [2, 0], [3, 0], [0, 1], [1, 1], [2, 1], [3, 1]]
B = len(data_num)
lam = 0.6
num_classes = 1000 if dataset == "imagenet" else 100

# Section C. Condition
es = []
# ls = [0, 2, 4, 6]
les = [-3, -1]
query_num = 15
columns = 4

# Section D. HeatMap Condition
fmt, border, cbar    = 0, True, False

# model_name = "PATCH"
device_type = 'cuda'
device = torch.device(device_type)

# 2. Load Data
dataset = JDataset(root_path, dataset, device)
samples, targets, imgs, label_names = dataset.getitems(data_num)
label_names.insert(0, f"{label_names[0]} : {lam} + {label_names[1]} : {1 - lam}")
label_names.insert(0, f"{label_names[0]} : {1-lam} + {label_names[1]} : {lam}")
if view[0]:
    drawImgPlot(dataset.unNormalize(samples), columns = label_names, col=len(samples))


model_name, model_path = "DEIT", "/hub_data/joonmyung/conference/2023ICCV/1.3.2/OURS:T:01/checkpoint_{}.pth"
model_number, ana_type, epoch = 1, "2023ICCV", 0

print("file list : ", os.listdir("/" + os.path.join(*model_path.split("/")[:-1])))

modelMaker = JModel(model_name, num_classes, model_path, model_number=model_number, ana_type = ana_type, device=device)
```

```python
es = [0, 20,40, 60, 80, 100, 120]
for e in es:
    model = modelMaker.getModel(e)[0]
    model.eval()
    outputs, output_12, loss_mask, loss_mae_1, loss_mae_2, mask_acc, mae_target, mae_target_mixed, mae_pred_1, mae_pred_2, indices, lam = model(samples)
    drawImgPlot(torch.stack([mae_target, mae_target_mixed, mae_target[indices], mae_pred_1, mae_pred_2], dim=1).reshape(-1, 3, 224, 224)[5:10], col=5);
```
