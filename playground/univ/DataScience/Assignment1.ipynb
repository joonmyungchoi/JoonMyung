```python
import numpy as np
import matplotlib.pyplot as plt

# 자유도 10의 카이제곱 분포를 생성하기 위해, 지수 분포를 10개 생성하여 합함
chi2 = np.sum(np.random.exponential(scale=1, size=(1000, 10)), axis=1)

# 평균과 표준편차 계산
mean = np.mean(chi2)
std = np.std(chi2)


print("평균 : ", mean)
print("표준편차 : ", std)

# 히스토그램 그리기
plt.hist(chi2, bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Chi-squared distribution with 10 degrees of freedom')
plt.show()
```
