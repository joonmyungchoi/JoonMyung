```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

lam = 3   
n = 10    # Number of years
max_k = n*lam   

probabilities = np.zeros(max_k+1)
for k in range(max_k+1):
    prob = 1 - poisson.cdf(k-1, n*lam)
    probabilities[k] = prob

k_values = np.arange(max_k+1)
plt.bar(k_values, probabilities)
plt.xlabel('Maximum number of earthquakes per year')
plt.ylabel('Probability')
plt.title(f'Distribution of maximum number of earthquakes in {n} years')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 10     # Sample size
N = 1000   # Number of simulations

sample_means = []
for i in range(N):
    X = np.random.uniform(0, 1, n)
    sample_mean = np.mean(X)
    sample_means.append(sample_mean)
plt.hist(sample_means, bins=20, density=True, alpha=0.5)

mu = 0.5 
sigma = np.sqrt(1/12)
x = np.linspace(0, 1, 1000)
y = 1/(sigma*np.sqrt(2*np.pi)) * np.exp(-(x-mu)**2/(2*sigma**2))
plt.plot(x, y, 'r-', lw=2)

plt.xlabel('Sample mean')
plt.ylabel('Density')
plt.title(f'Histogram of {N} sample means (n={n} observations each) and normal distribution')
plt.show()
```
