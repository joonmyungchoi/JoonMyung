```python
import numpy as np

# Transition matrix
P = np.array([[0.7, 0.4], [0.3, 0.6]])

# Initial state probability
pi = np.array([1, 0])

# Iterate until convergence
eps = 0.00005
diff = float('inf')
t = 0
while diff >= eps:
    pi_new = np.dot(pi, P)
    diff = np.max(np.abs(pi_new - pi))
    pi = pi_new
    t += 1

# Print the result
print(f"A에서 출발한 여행자는 '{t}스텝' 후 '{pi[0]:.6f}의 일정한 확률'을 갖게 된다.")
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt

k = 5
n = 10
p = 18/38
q = 1 - p

N = 1000
T_list = []


for i in range(N):
    money = k
    T = 0
    while money > 0 and money < n:
        if random.random() < p:
            money += 1
        else:
            money -= 1
        T += 1
    T_list.append(T)
    
T_mean = np.mean(T_list)
T_std = np.std(T_list)
print("T_mean:", T_mean)
print("T_std:", T_std)

plt.hist(T_list, bins=range(max(T_list)+2))
plt.xlabel("T")
plt.ylabel("Count")
plt.title("Histogram of T")
plt.show()
```
