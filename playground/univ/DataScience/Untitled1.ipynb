```python
# Number of trials and number of 'successes'
n_trials = 100
n_successes = 61

# Frequentist point estimate of the fairness of the coin
p_hat = n_successes / n_trials
print("Frequentist point estimate of the fairness of the coin:", p_hat)
```

```python
from scipy.stats import beta
import numpy as np
import matplotlib.pyplot as plt

# Prior parameters
alpha_prior = 1
beta_prior = 1

# Posterior parameters
alpha_post = n_successes + alpha_prior
beta_post = n_trials - n_successes + beta_prior

# Plot the prior and the posterior
x = np.linspace(0, 1, 100)
plt.plot(x, beta.pdf(x, alpha_prior, beta_prior), label='Prior')
plt.plot(x, beta.pdf(x, alpha_post, beta_post), label='Posterior')
plt.legend()
plt.title('Prior and Posterior distributions for the fairness of the coin')
plt.show()
```

```python
import numpy as np

# Simulate the testing of parts
np.random.seed(0)  # for reproducibility
n_parts = 1000  # total number of parts
p_defective = 0.05  # true probability of a part being defective
results = np.random.choice([0, 1], size=n_parts, p=[1-p_defective, p_defective])

# Frequentist estimation of the probability of a part being defective
n_defective = np.sum(results)
p_hat = n_defective / n_parts
print("Frequentist estimate of defect rate:", p_hat)
```

```python
from scipy.stats import beta
import matplotlib.pyplot as plt

# Prior parameters
alpha_prior = 1
beta_prior = 1

# Posterior parameters
alpha_post = n_defective + alpha_prior
beta_post = n_parts - n_defective + beta_prior

# Plot the prior and the posterior
x = np.linspace(0, 0.2, 100)
plt.plot(x, beta.pdf(x, alpha_prior, beta_prior), label='Prior')
plt.plot(x, beta.pdf(x, alpha_post, beta_post), label='Posterior')
plt.legend()
plt.title('Prior and Posterior distributions for the defect rate')
plt.show()
```

```python
# Prior probability of having cancer
p_cancer = 0.01  # Suppose 1% of the population has this type of cancer

# Test accuracy
p_pos_given_cancer = 0.9     # Sensitivity
p_pos_given_no_cancer = 0.1  # False positive rate

# You test positive. What's the probability you have cancer?
# Use Bayes' theorem to compute this.

# Prior probability of not having cancer
p_no_cancer = 1 - p_cancer

# Probability of testing positive (regardless of whether or not you have cancer)
p_pos = p_pos_given_cancer * p_cancer + p_pos_given_no_cancer * p_no_cancer

# Posterior probability of having cancer given that you tested positive
p_cancer_given_pos = p_pos_given_cancer * p_cancer / p_pos

print("Posterior probability of having cancer given a positive test:", p_cancer_given_pos)
```
